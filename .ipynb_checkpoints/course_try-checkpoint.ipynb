```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
udemy_data_raw = pd.read_csv("udemy_course_data.csv")
udemy_data_raw.head()
```

```python
udemy_title_raw = pd.read_csv("UdemyCleanedTitle.csv")
udemy_title_raw.head()
```

```python
udemy_data_raw.info()
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
from collections import Counter
```

```python
course_title = udemy_data_raw.course_title
word_list = []
corpus = []
for x in range(len(course_title)):
    clean_titles = re.sub("[^a-zA-Z0-9]", " ", course_title[x])
    clean_titles = clean_titles.lower()
    clean_titles = clean_titles.split()
    clean_titles = [WordNetLemmatizer().lemmatize(word, pos='v') for word in clean_titles if not word in set(stopwords.words('english'))]
    word_list.extend(clean_titles)
    corpus.append(" ".join(clean_titles))
count = pd.DataFrame(Counter(word_list).most_common(20))
corpus
count
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
cv = CountVectorizer()
token = cv.fit_transform(corpus)
x = token.toarray()
```

```python
df_cv_words = pd.DataFrame(data=x,columns=cv.get_feature_names())
df_cv_words
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cosine_sim_mat = cosine_similarity(token)
cosine_sim_mat
```

```python
durat_hours = []
for a in range(len(udemy_data_raw.content_duration)):
    cont_split = re.split('\s',udemy_data_raw.content_duration[a],maxsplit=1)
    if 'hours' in cont_split:
        durat_hours.append(float(cont_split[0]))
    elif 'mins' in cont_split:
        durat_hours.append(float(cont_split[0])/60)
    else:
        durat_hours.append(0)
len(durat_hours)
```

```python
udemy_data_raw['new_duration_hours']=durat_hours
```

```python
most_dur = udemy_data_raw.sort_values(by='new_duration_hours', ascending=False)
plt.figure(figsize=(12,6), dpi=100)
sns.barplot(x=most_dur.course_title[:15], y=most_dur.new_duration_hours[:15])
plt.xticks(rotation='vertical')
plt.show()
```
